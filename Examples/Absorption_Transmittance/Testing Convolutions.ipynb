```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from matplotlib import gridspec
import MATS
```

```python
wavenumbers = np.arange(-10, 10.005, 0.005)
VP = MATS.hapi.PROFILE_VOIGT(0, 0.1, 1, -0.01, wavenumbers, YRosen=0.0, Sw=1.0)

#ILS_dictionary = {'SINC2': {'function':MATS.hapi.SLIT_DIFFRACTION, resolution = 330.83*1e6/29979245800, width = 10}}

sinc = MATS.hapi.SLIT_DIFFRACTION(wavenumbers, 330.83*1e6/29979245800)
lorentz = MATS.hapi.SLIT_DISPERSION(wavenumbers, 133.56*1e6/29979245800)

#convolved = np.convolve(VP, sinc, mode = 'same')
#convolved = np.convolve(convolved, lorentz, mode = 'same')



con_waves, con_cross, i1, i2, slit = MATS.hapi.convolveSpectrumSame(
    wavenumbers,
    VP,
    Resolution=330.83*1e6/29979245800,
    AF_wing=10,
    SlitFunction=MATS.hapi.SLIT_DIFFRACTION,
    Wavenumber=None)

con_waves, con_cross, i1, i2, slit = MATS.hapi.convolveSpectrumSame(
    wavenumbers,
    con_cross,
    Resolution=133.56*1e6/29979245800,
    AF_wing=10,
    SlitFunction=MATS.hapi.SLIT_DISPERSION,
    Wavenumber=None)



plt.plot(wavenumbers, VP)
plt.plot(con_waves, con_cross)

```
